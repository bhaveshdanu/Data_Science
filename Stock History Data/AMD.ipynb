```python
import yfinance
```

##**TESLA Stocks**

```python
df=yfinance.download('TSLA',period='3d',interval='2m')
```

```python
df.head()
```

##**Tesla Graph**

```python
import matplotlib.pyplot as ply
%matplotlib inline
```

```python
x=df.index
y1=df['Volume']
y2=df['Adj Close']
fig,ax1=ply.subplots()
ax2=ax1.twinx()
curve1=ax1.plot(x,y1,label='Volume',color='r')
curve2=ax2.plot(x,y2,label='Close',color='b')
ply.plot()
ply.show()
```
